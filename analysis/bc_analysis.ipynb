```python
import math
dfbc2 = pd.read_csv("../dataset/Mexico/bc_data/BDmortalidadCAMAbase.csv")
dfbc2 = process_df(dfbc2,2)
dfbc2['year'] = dfbc2['certification_year']
dfbc2['age_group'] = dfbc2['age_group'].apply(update_age_group)
df_smr = pd.merge(dfbc2, df_asmr, on=asmrcols)
df_smr['observed_deaths'] = 1
aggrcol = ['state_code','municipality_code']
if isinstance(aggrcol, str):
    aggrcol = [aggrcol]
if isinstance(asmrcols, str):
    asmrcols = [asmrcols]
cols_for_aggregation = asmrcols + aggrcol
df_smra = df_smr.groupby(cols_for_aggregation).agg({
    'observed_deaths': 'sum',
    'population': 'mean',
    'asmr': 'mean'
    # 'state_code': 'mean',
    # 'municipality_code': 'mean'
})
# aggrcol='age_group'
df_smra['expected_deaths'] = df_smra['population'] * df_smra['asmr']
df_smr2 = df_smra.groupby(cols_for_aggregation).agg({
    'observed_deaths': 'sum',
    'expected_deaths': 'sum',
    'population': 'sum'
}).reset_index()
df_smr2['expected_deaths'].apply(math.ceil)
df_smr2['SMR'] = df_smr2['observed_deaths'] / df_smr2['expected_deaths']
```
